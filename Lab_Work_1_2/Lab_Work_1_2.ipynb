#### Q4 - Fetching Data from an API

```python
# Fetch from API

load_dotenv()
api_key = os.environ.get("API_KEY")

url = "https://rest.coincap.io/v3/markets"
headers = {"Authorization": f"Bearer {api_key}"}

response = requests.get(url, headers=headers)
```

```python
# Converted to DF

if response.status_code == 200:
    data = response.json()
    data = data['data']

    coins = pd.DataFrame(data)
else:
    print("Error Code: ", response.status_code)
```

```python
# Display first 5 rows, column name, and shape

print(f'First 5 Rows: \n{coins.head()}\n')

print(f'Student Columns Names: {coins.columns}\n')

print(f'Shape: {coins.shape}\n')
```

```python
# Filtered (exchangeId = binance and coinex)

binance_coinex = pd.DataFrame(coins[(coins['exchangeId'] == 'binance') | (coins['exchangeId'] == 'coinex')])
binance_coinex.head()
```

#### Q5 - Data Cleaning Process

```python
# Load CSV Dataset

uncleaned_data = pd.read_csv('uncleaned_data.csv')
```

```python
# Display first 5 rows, column name, and shape

print(f'First 5 Rows: \n{uncleaned_data.head()}\n')

print(f'Student Columns Names: {uncleaned_data.columns}\n')

print(f'Shape: {uncleaned_data.shape}\n')
```

```python
# Before Cleaning

print(f'Uncleaned Data: \n{uncleaned_data.head()}\n')
```

```python
print('Info:')
print(uncleaned_data.info())
```

```python
print('\nDescribe:')
print(uncleaned_data[['age', 'marks']].describe())
```

```python
print('Null Value Count:')
print(uncleaned_data.isnull().sum())
```

```python
' As you can see, the dataset has missing values, outliers(marks and age), not standardize(department), and date format. '
```

```python
uncleaned_data
```

```python
# Cleaning Process

clean_data = uncleaned_data

# Handled Missing Values
clean_data.fillna({'age': clean_data['age'].median(),'marks': clean_data['marks'].median()},inplace=True)
clean_data.dropna(inplace=True)
```

```python

# Handled Outliers

def outlier(colname,data):
    q1 = data[colname].quantile(0.25)
    q3 = data[colname].quantile(0.75)
    iqr = q3 - q1

    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr

    new_data = data[(data[colname] >= lower) & (data[colname] <= upper)]

    return new_data

clean_data = outlier('age', clean_data)
clean_data = outlier('marks', clean_data)

clean_data
```

```python
# Handled Standardize

clean_data['department'] = clean_data['department'].str.upper()
clean_data
```

```python
# Handled Date Format

clean_data['joining_date'] = pd.to_datetime(clean_data['joining_date'],format='mixed')
clean_data
```

```python
# Exported Clean CSV

clean_data.to_csv('cleaned_data.csv', index=False)
```
