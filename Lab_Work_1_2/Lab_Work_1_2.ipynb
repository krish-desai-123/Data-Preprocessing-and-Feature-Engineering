```python
import pandas as pd
```

#### Q1 - Working with CSV File

```python
# Load CSV Dataset

students = pd.read_csv('students.csv')
```

```python
# Display first 5 rows, column name, and shape

print(f'First 5 Rows: \n{students.head()}\n')

print(f'Student Columns Names: {students.columns}\n')

print(f'Shape: {students.shape}\n')
```

```python
# Filtered (marks > 70)

mark_gte_70 = students[students['marks']>70]
print(f'Filtered (marks > 70): \n{mark_gte_70}')
```

```python
# Save the CSV

mark_gte_70.to_csv('filtered_students.csv', index=False)
```

#### Q2 - Working with JSON File

```python
# Load JSON Dataset

employee = pd.read_json('employee.json')
```

```python
# Display first 5 rows, column name, and shape

print(f'First 5 Rows: \n{employee.head()}\n')

print(f'Student Columns Names: {employee.columns}\n')

print(f'Shape: {employee.shape}\n')
```

```python
# Filtered (department == IT)

IT = employee[employee['department']=='IT']
print(f'Filtered (department = IT): \n{IT}')
```

```python
# Save the JSON

IT.to_json('filtered_employee.json', orient='records', indent=2)
```
