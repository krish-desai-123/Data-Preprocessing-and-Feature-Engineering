```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Created dataset

df = pd.DataFrame({
    'OrderID' : [1,2,3,4,5,6],
    'OrderDate': ['2026-01-05','2026-01-14','2026-01-21','2026-01-22','2026-02-05','2026-02-09',],
    'DeliveryDate': ['2026-01-10','2026-01-20','2026-01-27','2026-01-25','2026-02-9','2026-02-16',]
})
display(df)
```

```python
df.info()
```

```python
# Convert to datetime format

df['OrderDate'] = pd.to_datetime(df['OrderDate'])
df['DeliveryDate'] = pd.to_datetime(df['DeliveryDate'])
display(df)
```

```python
df.info()
```

```python
# Extract

df['OrderYear'] = df['OrderDate'].dt.year
df['OrderMonth'] = df['OrderDate'].dt.month
df['OrderDay'] = df['OrderDate'].dt.day

df['DeliveryDuration'] = df['DeliveryDate'].dt.day - df['OrderDate'].dt.day
display(df)
```

```python
df['is_weekday'] = ~df['OrderDate'].dt.weekday.isin([5,6])
df['is_weekend'] =  df['OrderDate'].dt.weekday.isin([5,6])
```

```python
# Bar Chart

sns.countplot(data=df, x='OrderMonth')
plt.xticks(ticks=[0,1], labels=['January','February'])
plt.ylabel('Order Count')
plt.xlabel('Order Month')
plt.show()
```
