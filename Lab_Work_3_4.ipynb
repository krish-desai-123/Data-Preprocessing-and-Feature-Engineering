```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, Normalizer, MinMaxScaler, MaxAbsScaler, RobustScaler
```

#### Q1 - Standardization

```python
df = pd.DataFrame({
    'height': [50, 52, 38, 44, 36, 48, 42, 40, 55, 30],
    'weight': [140, 130, 135, 145, 162, 175, 140, 155, 158, 165],
})

ss = StandardScaler()
df[['height_scaled','weight_scaled']] = ss.fit_transform(df)

print('Before')
print(f'Height Mean: {df['height'].mean().round(2)}, Weight: {df['weight'].mean().round(2)}')
print(f'Height Std: {df['height'].std().round(2)}, Weight: {df['weight'].std().round(2)}')

print('After')
print(f'Height Mean: {df['height_scaled'].mean().round(2)}, Weight: {df['weight_scaled'].mean().round(2)}')
print(f'Height Std: {df['height_scaled'].std().round(2)}, Weight: {df['weight_scaled'].std().round(2)}')
```

#### Q2 - Normalization

```python
df = pd.DataFrame({
    'user':     ['Aarav','Priya','Ravi','Sneha','Karan'],
    'biryani':  [10, 2, 0, 8, 5],
    'pizza':    [2, 8, 1, 3, 6],
    'dosa':     [5, 1, 9, 2, 4],
    'burger':   [1, 5, 2, 7, 3],
})
cols = ['biryani','pizza','dosa','burger']
l2 = Normalizer(norm='l2')

df_l2 = df.copy()
df_l2[cols] = l2.fit_transform(df[cols])

print('Normalized: ')
display(df_l2.round(2))

print('Verify:')
row_norms = np.sqrt((df_l2[cols]**2).sum(axis=1))
print(row_norms.values)
```

#### Q3 - Min-Max Scaling

```python
np.random.seed(9009)
df = pd.DataFrame({
    'salary' : np.random.randint(20000,100000,20)
})

mm = MinMaxScaler()
df_mm = pd.DataFrame(mm.fit_transform(df), columns=df.columns)

print('MinMax Scaled: ')
display(df_mm.round(2).head())
```

#### Q4 - Max-Abs Scaling

```python
df = pd.DataFrame({
    'balance' : np.random.randint(-20000,20000,20)
})

ma = MaxAbsScaler()
df_ma = pd.DataFrame(ma.fit_transform(df), columns=df.columns)

print('MaxAbs Scaled: ')
display(df_ma.round(2).head())
```

#### Q5 - Robust Sacling

```python
df = pd.DataFrame({
    'income_lakh':  [8, 12, 6, 15, 9, 11, 10, 14, 0.5, 200]
})

rs = RobustScaler()
df_rs = pd.DataFrame(rs.fit_transform(df), columns=df.columns)
print('Robust Scaled: ')
display(df_rs.round(2))
```
