```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import KBinsDiscretizer, Binarizer
```

#### Q1 - Binning Discretization

```python
# Created dataset

np.random.seed(9009)
df = pd.DataFrame({
    'Age' : np.random.randint(18,70, 100)
})
display(df.head())
```

```python
# Apply

df['Age_group'] = pd.cut(
    df['Age'],
    bins=[17,30,50,70],
    labels=['Young','Middle','Senior'],
    right=True
)
display(df.head(10))
```

```python
# Counts

print(df['Age_group'].value_counts().sort_index())
```

#### Q2 - Binarization

```python
# Created dataset

df = pd.DataFrame({
    'Income' : np.random.randint(20000,100000, 100)
})
display(df.head())
```

```python
# Apply

inc_bin = Binarizer(threshold=50000)
df['Income_bin'] = inc_bin.fit_transform(df[['Income']])
display(df.head(10))
```

```python
print(df['Income_bin'].value_counts().sort_index())
```

#### Q3 - Quantile Binning

```python
# Created dataset

df = pd.DataFrame({
    'Income' : np.random.lognormal(10,0.5,100)
})
display(df.head())
```

```python
# Apply

inc_bin = KBinsDiscretizer(n_bins = 4, encode = 'ordinal', strategy='quantile',quantile_method='averaged_inverted_cdf')
df['Income_bin'] = inc_bin.fit_transform(df[['Income']])
display(df.head(10))
```

```python

print(df['Income_bin'].value_counts().sort_index())
```

#### Q4 - K-Means Binning

```python
# Created dataset

df = pd.DataFrame({
    'SpendingScore': np.random.rand(100)
})
display(df.head())
```

```python
# Apply

ss_bin = KBinsDiscretizer(n_bins = 4, encode = 'ordinal', strategy='kmeans')
df['SpendingScore_bin'] = ss_bin.fit_transform(df[['SpendingScore']])
display(df.head(10))
```

```python
print(df['SpendingScore_bin'].value_counts().sort_index())
```
