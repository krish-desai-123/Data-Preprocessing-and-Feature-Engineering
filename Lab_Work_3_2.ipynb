```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder, LabelEncoder, OneHotEncoder
```

#### Q1 - Data Ordinal Encoding

```python
# Created dataset

df = pd.DataFrame({
    'applicant': ['Aarav', 'Priya', 'Ravi', 'Sneha', 'Karan',
                  'Divya', 'Mohit', 'Ananya'],
    'education': ['Graduate',      'Post-Graduate', 'School',
                  'Graduate',      'Post-Graduate', 'School',
                  'Graduate',      'Post-Graduate'],

})
display(df)
```

```python
# Encode

ordenc = OrdinalEncoder(
    categories=[['School', 'Graduate', 'Post-Graduate']]
)

df['education_enc'] = ordenc.fit_transform(df[['education']]).astype(int)
display(df)
```

#### Q2 - Label Encoding

```python
# Create dataset

df = pd.DataFrame({
    'driver_id': [1, 2, 3, 4, 5, 6, 7, 8],
    'city':      ['Mumbai', 'Delhi', 'Pune', 'Mumbai',
                  'Chennai', 'Delhi', 'Pune', 'Chennai'],
})
display(df)
```

```python
# Encoding

labenc = LabelEncoder()

df['city_enc'] = labenc.fit_transform(df['city'])+1
display(df)
```

#### Q3 - One-Hot Encoding

```python
# Created dataset

df = pd.DataFrame({
    'emp_id':  [1001, 1002, 1003, 1004, 1005, 1006, 1007, 1008],
    'department':   ['HR',       'IT',     'Finance',     'Sales',
                  'IT',      'IT',      'Finance',     'Sales'],

})
display(df)
```

```python
# Encoding using OneHotEncoder

ohenc = OneHotEncoder(
    drop='first',
    handle_unknown='ignore',
    sparse_output=False
)

col = ['department']
enc_array = ohenc.fit_transform(df[col])

col_names = ohenc.get_feature_names_out(col)

df_ohe = pd.DataFrame(enc_array, columns=col_names)
display(df_ohe)
```

```python
# Encoding using pd.get_dummies()

df_pd = pd.get_dummies(df['department'], drop_first=True, prefix='department')
display(df_pd)
```
